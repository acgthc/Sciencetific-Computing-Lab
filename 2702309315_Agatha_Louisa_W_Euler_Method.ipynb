```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exactn Solution \ for Simpe ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters for the ODE y' = x^2 - 4y
f = lambda x, y: x**2 - 4*y
h = 0.1  # Step size
x = np.arange(0, 1 + h, h)  # Numerical grid
y0 = 1  # Initial Condition

# Explicit Euler Method
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Perform one step of the Explicit Euler Method to find y(0.1)
y0_1 = y[1]

# Output the result
print(f"y(0.1) using the explicit Euler method is approximately: {y0_1}")

# Plot the solution
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Approximate Solution')
plt.title("Approximate Solution for ODE y' = x^2 - 4y")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda y: np.sin(y) # ODE
h = 0.1 # Step size
x = np.arange(0, 0.5 + h, h) # Numerical grid
y0 = 1 # Initial Condition

# Explicit Euler Method
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h*f(y[i])

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.title('Approximate Solution for y\' = sin(y)')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
